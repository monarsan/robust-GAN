```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from gan import gan as GAN
import seaborn as sns
```

```python
gan = GAN(10, 0.1)
gan.dist_init('sigma', 0, 5)
gan.data_init(1000, 3)
gan.model_init(D_init_option='exp')
gan.optimizer_init(0.05, 0.1, 0.04, 1e-5, 0, 1, 1)
gan.fit(10)

plt.subplot(1, 2, 1)
plt.plot(gan.l2_loss)
plt.subplot(1, 2, 2)
plt.plot(gan.D_contami_record, label='contami')
plt.plot(gan.D_target_record, label='target')
plt.plot(gan.D_data_record)
plt.plot(gan.D_z_record, label='z')
plt.legend()
plt.plot(np.array(gan.D_z_record) - np.array(gan.D_data_record))
```
