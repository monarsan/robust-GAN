```python
import numpy as np
import matplotlib.pyplot as plt
from gan import gan as GAN
import seaborn as sns
```

```python
gan = GAN(data_dim=2, eps=0.1)
gan.dist_init(setting='mu', true_mean=5, out_mean=0)
gan.data_init(data_size=1000, mc_ratio=3)
gan.model_init()
gan.optimizer_init(lr_d=1, lr_g=0.05, decay_par=0.4,
                   reg_d=6e-5, reg_g=5e-5, update_D_iter=5, is_mm_alg=False)
gan.fit(optim_iter=1000, verbose=True)
plt.plot(gan.objective)
# plt.plot(gan.D_record)
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 3, 1)
plt.plot(gan.l2_loss)
plt.xlabel('optim step')
plt.ylabel('l2 loss')
plt.title(f'data dim is {gan.data_dim}')
plt.subplot(1, 3, 2)
plt.plot(np.array(gan.G_record))
plt.xlabel('optim step')
plt.ylabel('estimated value of mean')
plt.title(f'data dim is {gan.data_dim}')

plt.subplot(1,3,3)
plt.plot(np.array(gan.D_contami_record), label='cont')
plt.plot(np.array(gan.D_z_record), label='z')
plt.plot(np.array(gan.D_data_record), label='target')
plt.xlabel('optim step')
plt.ylabel('mean of the output of Discriminator')
plt.legend()
plt.subplots()
plt.plot(gan.bias_record)
```
