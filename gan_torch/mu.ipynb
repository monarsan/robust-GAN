```python
from mu import Mu
import numpy as np
import matplotlib.pyplot as plt


data_dim = 50
for _ in range(1):
    gan = Mu(data_dim, 0.1, 'cpu')
    true_mean = np.zeros(data_dim)
    out_mean = np.ones(data_dim) * 5
    gan.dist_init(true_mean, out_mean)
    gan.data_init(1000, 25)
    gan.model_init(D_model='quadratic')
    gan.optimizer_init(0.03, 0.003, 5, 1, weight_decay_d=0.025, weight_decay_g=2e-5)
    gan.fit(500)
    plt.plot(gan.mean_err_record, label='mean')
```
